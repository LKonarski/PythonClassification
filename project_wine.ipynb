```python
import pandas as pd

df = pd.read_csv('winedataset.csv')
df = df.fillna(0)
df
```

```python
X = df.drop(columns=['type'])
X
```

```python
df.insert(0, 'color', df['type'].map(lambda x: x == 'white').astype(int))
df
```

```python
y = df['color']
y
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
model = DecisionTreeClassifier(max_depth=4)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

score = accuracy_score(y_test, y_pred)
score
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(50, 35))
tree.plot_tree(model, feature_names=X.columns, class_names=['white', 'red'], filled=True)
plt.show()
```

```python
models = {}

from sklearn.linear_model import LogisticRegression
models['Logistic Regression'] = LogisticRegression()

from sklearn.svm import LinearSVC
models['Support Vector Machines'] = LinearSVC()

from sklearn.tree import DecisionTreeClassifier
models['Decision Trees'] = DecisionTreeClassifier()

from sklearn.ensemble import RandomForestClassifier
models['Random Forest'] = RandomForestClassifier()

from sklearn.naive_bayes import GaussianNB
models['Naive Bayes'] = GaussianNB()

from sklearn.neighbors import KNeighborsClassifier
models['K-Nearest Neighbor'] = KNeighborsClassifier()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy, precision, recall = {}, {}, {}

for key in models.keys():
    
    models[key].fit(X_train, y_train)

    y_pred = models[key].predict(X_test)
    
    accuracy[key] = accuracy_score(y_pred, y_test)
    precision[key] = precision_score(y_pred, y_test)
    recall[key] = recall_score(y_pred, y_test)
```

```python
import pandas as pd

df_model = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])
df_model['Accuracy'] = accuracy.values()
df_model['Precision'] = precision.values()
df_model['Recall'] = recall.values()

df_model
```

```python
ax = df_model.plot.bar(color=['green', 'red', 'purple'])
ax.legend(
    ncol=len(models.keys()), 
    bbox_to_anchor=(0, 1), 
    loc='lower left', 
    prop={'size': 15}
)
plt.tight_layout()
```

```python
import numpy as np

np.random.seed(7)
df = pd.read_csv('winedataset.csv', usecols=['alcohol', 'pH']).sample(20)
df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 15), nrows=3)

# ---
ax[0].scatter(df["alcohol"], df["pH"], label='Real data')
ax[0].set_xlabel("Alcohol", fontsize=15)
ax[0].set_ylabel("PH", fontsize=15)
ax[0].set_title("Alcohol and pH of Selected Wines", fontsize=18)

# ---
ax[1].scatter(df["alcohol"], df["pH"], label='Real data')
a, b = np.polyfit(df["alcohol"], df["pH"], 1)
ax[1].plot(df["alcohol"], a*df["alcohol"]+b, 'r', label='Approximation y=a*x+b')
ax[1].set_title(f"Line of Best Fit: $y$={round(a, 4)}$x${round(b, 1)}", fontsize=18)
ax[1].set_xlabel("Alcohol", fontsize=15)
ax[1].set_ylabel("Quality", fontsize=15)
ax[1].legend()

# ---
ax[2].scatter(df["alcohol"], df["pH"], label='Real data', alpha=0.3)
a, b = np.polyfit(df["alcohol"], df["pH"], 1)
plt.plot(df["alcohol"], a*df["alcohol"]+b, 'r', label='Approximation $y$=a*$x$+b')
ax[2].set_xlabel("Alcohol", fontsize=15)
ax[2].set_ylabel("PH", fontsize=15)
ax[2].legend()
ax[2].set_title("Determining pH for Alcohol", fontsize=18)

fig.subplots_adjust(hspace=0.5)
```
